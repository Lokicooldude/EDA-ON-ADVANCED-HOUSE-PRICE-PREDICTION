```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
pd.pandas.set_option('display.max_columns',None)  ##to display all columns
```

```python
df1=pd.read_csv('train.csv')
```

```python
df1.head()
```

```python
df1.shape
```

```python
#missing values
fetures_with_na=[features for features in df1.columns if(df1[features].isnull().sum()>1)]

for features in fetures_with_na:
    print(features,np.round(df1[features].isnull().mean(),4)*100,'% missing values')
```

```python
#before we drop the missing values we have to check the relationship between missingvalues and dependent variable
for features in fetures_with_na:
    data=df1.copy()
    data[features]=np.where(df1[features].isnull(),1,0)
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
```

```python
numerical_features=[features for features in df1.columns if(df1[features].dtype!='O')]
print(numerical_features)
print(len(numerical_features))
```

```python
df1[numerical_features].head()
```

```python
year_feature=[feature for feature in numerical_features if ('Yr' in feature or 'Year' in feature)]


print(year_feature)
```

```python
for feature in year_feature:
    print(feature,df1[feature].unique())
```

```python
df1.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Sale price')
plt.title('house price vs year sold')
```

```python
year_feature
```

```python
for feature in year_feature:
    if(feature!='YrSold'):
        data=df1.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.show()
        
        
```

```python
#discretefeatures
discrete_features=[feature for feature in numerical_features if((len(df1[feature].unique())<25)and(feature not in year_feature+['Id']))]
len(discrete_features)
```

```python
discrete_features
```

```python
df1[discrete_features].head()
```

```python
for features in discrete_features:
    data=df1.copy()
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.xlabel(features)
    plt.ylabel('Sale Price')
    plt.title(features)
    plt.show()
```

```python
#continuosvariables
continuous_features=[features for features in numerical_features if features not in discrete_features+year_feature+['Id']]
continuous_features
```

```python
len(continuous_features)
```

```python
#creating histogram to understand the distribution
for feature in continuous_features:
    data=df1.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title(feature)
    plt.show()
    
```

```python
data=df1.copy()
for feature in continuous_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        data[feature].hist(bins=25)
        plt.xlabel('SalePrice')
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
#outliers

for feature in continuous_features:
    data=df1.copy()
    if 0 in data[feature].unique():
        
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()

       
```
